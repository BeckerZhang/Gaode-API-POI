```python
import requests

polygon = '120.60961,31.329655|120.629437,31.33265;120.634152,31.332134|120.63657,31.323252|120.634757,31.32129|120.642615,31.292677|120.61638,31.289165|120.610456,31.288028|120.609973,31.289681|120.607675,31.299186|120.605137,31.30621|120.601993,31.309412|120.600301,31.324905|120.603202,31.326144|120.605983,31.329139|120.60961,31.329655'
types = '090000|090100|090101|090102|090200|090201|090202|090203|090204|090205|090206|090207|090208|090209|090210|090211|090300|090400|090500'
key = 'Your_Key'
keyword = '医院'

url = 'https://restapi.amap.com/v5/place/polygon?key={}&polygon={}&keywords={}&types={}&page_size=25&page_num={}'
page_num = 1  # Start with the first page
all_results = []

while True:
    current_url = url.format(key, polygon, keyword, types, page_num)
    print(f"Requesting page {page_num}...")
    req = requests.get(current_url)

    if req.status_code == 200:
        data = req.json()
        
        # Check if there are results on the current page
        if 'pois' in data and data['pois']:
            all_results.extend(data['pois'])
            page_num += 1
        else:
            # No more results, break out of the loop
            break
    else:
        print(f"Error: {req.status_code} - {req.text}")
        break

# Now 'all_results' contains the data from all pages
all_results
```

```python
import json
import pandas as pd
data_list = all_results

df = pd.DataFrame(data_list, columns=['name', 'location', 'type'])

# Extract latitude and longitude from the 'location' column
df[['longitude', 'latitude']] = df['location'].str.split(',', expand=True)

# Drop the original 'location' column
df = df.drop('location', axis=1)

# Print the DataFrame
print(df)
```

```python
import math

def gcj02_to_wgs84(lng, lat):
    """
    GCJ02(火星坐标系)转WGS84
    :param lng: 火星坐标系的经度
    :param lat: 火星坐标系纬度
    :return: 转换后的WGS84坐标系经纬度
    """
    if out_of_china(lng, lat):
        return [lng, lat]
    dlat = _transformlat(lng - 105.0, lat - 35.0)
    dlng = _transformlng(lng - 105.0, lat - 35.0)
    radlat = lat / 180.0 * math.pi
    magic = math.sin(radlat)
    magic = 1 - 0.00669342162296594323 * magic * magic
    sqrtmagic = math.sqrt(magic)
    dlat = (dlat * 180.0) / ((6378245.0 * (1 - 0.00669342162296594323)) / (magic * sqrtmagic) * math.pi)
    dlng = (dlng * 180.0) / (6378245.0 / sqrtmagic * math.cos(radlat) * math.pi)
    mglat = lat + dlat
    mglng = lng + dlng
    return [lng * 2 - mglng, lat * 2 - mglat]

def _transformlat(lng, lat):
    ret = -100.0 + 2.0 * lng + 3.0 * lat + 0.2 * lat * lat + 0.1 * lng * lat + 0.2 * math.sqrt(math.fabs(lng))
    ret += (20.0 * math.sin(6.0 * lng * math.pi) + 20.0 * math.sin(2.0 * lng * math.pi)) * 2.0 / 3.0
    ret += (20.0 * math.sin(lat * math.pi) + 40.0 * math.sin(lat / 3.0 * math.pi)) * 2.0 / 3.0
    ret += (160.0 * math.sin(lat / 12.0 * math.pi) + 320.0 * math.sin(lat * math.pi / 30.0)) * 2.0 / 3.0
    return ret

def _transformlng(lng, lat):
    ret = 300.0 + lng + 2.0 * lat + 0.1 * lng * lng + 0.1 * lng * lat + 0.1 * math.sqrt(math.fabs(lng))
    ret += (20.0 * math.sin(6.0 * lng * math.pi) + 20.0 * math.sin(2.0 * lng * math.pi)) * 2.0 / 3.0
    ret += (20.0 * math.sin(lng * math.pi) + 40.0 * math.sin(lng / 3.0 * math.pi)) * 2.0 / 3.0
    ret += (150.0 * math.sin(lng / 12.0 * math.pi) + 300.0 * math.sin(lng / 30.0 * math.pi)) * 2.0 / 3.0
    return ret

def out_of_china(lng, lat):
    return not (lng > 73.66 and lng < 135.05 and lat > 3.86 and lat < 53.55)
```

```python
import pandas as pd

# Assuming you already have the functions gcj02_to_wgs84 and the DataFrame df

# Apply the conversion to each row in the DataFrame
df[["longitude_wgs84", "latitude_wgs84"]] = df.apply(lambda row: pd.Series(gcj02_to_wgs84(float(row['longitude']), float(row['latitude']))), axis=1)

df
```
